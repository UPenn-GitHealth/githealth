```python
import plotly.express as px

# Filter out NaN values in the 'affiliation' column
filtered_final_df = final_df.dropna(subset=['affiliation'])

# Count unique 'user_id' values for each affiliation
affiliation_counts = filtered_final_df['affiliation'].value_counts().reset_index()
affiliation_counts.columns = ['Affiliation', 'Count']

# Create an interactive histogram using Plotly
fig = px.bar(affiliation_counts, x='Affiliation', y='Count', title='Count of Unique User IDs for Each Affiliation')

# Update layout to enable zoom in/out functionality and fit to screen
fig.update_layout(
    xaxis=dict(title='Affiliation'),
    yaxis=dict(title='Count of Unique User IDs'),
    autosize=True,  # Adjusts the figure size to fit the screen
    margin=dict(l=50, r=50, t=50, b=50),
    template='plotly_white'
)

# Save the interactive figure as an HTML file
fig.write_html('affiliation_histogram.html')
```
