```python
df['Start_year'] = pd.to_datetime(df.Start_time).dt.year
df['Start_month'] = pd.to_datetime(df.Start_time).dt.month
df['Start_day'] = pd.to_datetime(df.Start_time).dt.day
df['End_year'] = pd.to_datetime(df.End_time).dt.year
df['End_month'] = pd.to_datetime(df.End_time).dt.month
df['End_day'] = pd.to_datetime(df.End_time).dt.day

# df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
```

```python
df.head()
```

```python
idf = df.interactive()
year_slider_min = pn.widgets.IntSlider(name='Year slider min', start=2018, end=2024, step=1, value=2018)
year_slider_max = pn.widgets.IntSlider(name='Year slider max', start=2018, end=2024, step=1, value=2023)
```

```python
pulls_per_repo = (
    idf.groupby(['Repository'])
    .size()
    .reset_index(name='Pull_Request_Count')
    .set_index(['Repository'])
    .sort_values(by=['Pull_Request_Count'], ascending=False)
)
pulls_per_repo

# show the table
pn.widgets.Tabulator(pulls_per_repo)
```

```python
# Make Pull Request Counter
pulls_per_repo_slider = pn.widgets.IntSlider(name='Pull Request Counts', start=0, end=6000, step=50, value=0)
pulls_per_repo_slider

# Slide the pull request counter to display chart accordingly
pulls_per_repo_pipeline = (
    pulls_per_repo[pulls_per_repo.Pull_Request_Count >= pulls_per_repo_slider]
    .sort_values(by=['Pull_Request_Count'], ascending=False)
)

# Use panel to display bar chart by passing Organisation and Counts
pulls_per_repo_bar_plot = pulls_per_repo_pipeline.hvplot(kind='bar', 
                                                     x='Repository', 
                                                     y='Pull_Request_Count', 
                                                     width=600, 
                                                     height=600,
                                                     rot=90,
                                                     title='Number of Pulls per Repository')
pulls_per_repo_bar_plot
```

```python
# Number of Open Pull Requests
open_pulls_per_repo = (
    # only include open pull requests
    idf[idf['PR_Status'] == 'open']
    .groupby(['Repository'])
    .size()
    .reset_index(name='Open_Pull_Request_Count')
    .set_index(['Repository'])
    .sort_values(by=['Open_Pull_Request_Count'], ascending=False)
)
pulls_per_repo

# show the table
pn.widgets.Tabulator(pulls_per_repo)

# Make Open Pull Request Counter
open_pulls_per_repo_slider = pn.widgets.IntSlider(name='Open Pull Request Counts', start=0, end=300, step=5, value=0)
open_pulls_per_repo_slider

# Slide the open pull request counter to display chart accordingly
open_pulls_per_repo_pipeline = (
    open_pulls_per_repo[open_pulls_per_repo.Open_Pull_Request_Count >= open_pulls_per_repo_slider]
    .sort_values(by=['Open_Pull_Request_Count'], ascending=False)
)

# Use panel to display bar chart by passing Organisation and Counts

open_pulls_per_repo_bar_plot = open_pulls_per_repo_pipeline.hvplot(kind='bar',
                                                                     x='Repository',
                                                                     y='Open_Pull_Request_Count',
                                                                     width=600,
                                                                     height=600,
                                                                     rot=90,
                                                                     title='Number of Open Pulls per Repository')
open_pulls_per_repo_bar_plot
```

```python
# chart the number of pull requests per year-month

# first merge the year and month into the same column in the format YYYY-MM (include a 0 at the start of the month if it's a single digit month), then group by that column

df['Start_month'] = df['Start_month'].apply(lambda x: '{0:0>2}'.format(x))
df['Start_year_month'] = df['Start_year'].astype(str) + '-' + df['Start_month'].astype(str)



pulls_per_month = (
    idf.groupby(['Start_year_month'])
    .size()
    .reset_index(name='Monthly_Pulls')
    .set_index(['Start_year_month'])
    .sort_values(by=['Start_year_month'], ascending=True)
)
pulls_per_month

# merge the year and month into the same column in the format YYYY-MM
# pulls_per_month['Start_year_month'] = pulls_per_month['Start_year'].astype(str) + '-' + pulls_per_month['Start_month'].astype(str)

pn.widgets.Tabulator(pulls_per_month)



# # graph the table

pulls_per_month_plot = pulls_per_month.hvplot(kind='bar',
                                                x='Start_year_month',
                                                y='Monthly_Pulls',
                                                width=600,
                                                height=600,
                                                rot=90,
                                                title='Number of Pulls per Month')
pulls_per_month_plot
```
