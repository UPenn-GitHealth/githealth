```python
df['Start_year'] = pd.to_datetime(df.Start_time).dt.year
df['Start_month'] = pd.to_datetime(df.Start_time).dt.month
df['Start_day'] = pd.to_datetime(df.Start_time).dt.day
df['End_year'] = pd.to_datetime(df.End_time).dt.year
df['End_month'] = pd.to_datetime(df.End_time).dt.month
df['End_day'] = pd.to_datetime(df.End_time).dt.day

# df['Month'] = df['Month'].apply(lambda x: calendar.month_abbr[x])
```

```python
idf = df.interactive()
year_slider_min = pn.widgets.IntSlider(name='Year slider min', start=1957, end=2020, step=1, value=1957)
year_slider_max = pn.widgets.IntSlider(name='Year slider max', start=1957, end=2020, step=1, value=2000)
```

```python
pulls_per_repo = (
    idf.groupby(['Repository'])
    .size()
    .reset_index(name='Pull_Request_Count')
    .set_index(['Repository'])
    .sort_values(by=['Pull_Request_Count'], ascending=False)
)
pulls_per_repo

# show the table
pn.widgets.Tabulator(pulls_per_repo)
```

```python
# # Make Launch Counter
# launch_per_org_slider = pn.widgets.IntSlider(name='Launch Counts', start=1, end=1777, step=1, value=1)
# launch_per_org_slider

# # Slide the launch counter to display chart accordingly
# launch_per_org_pipeline = (
#     launch_per_org[launch_per_org.Counts >= launch_per_org_slider]
#     .sort_index(ascending=False)
#     .reset_index(drop=True)
# )

# # Use panel to display bar chart by passing Organisation and Counts
# launch_per_org_bar_plot = launch_per_org_pipeline.hvplot(kind='bar', 
#                                                      x='Organisation', 
#                                                      y='Counts', 
#                                                      width=1000, 
#                                                      height=400,
#                                                      rot=90,
#                                                      title='Number of Launch per Organisation')
# launch_per_org_bar_plot

# Make Pull Request Counter
pulls_per_repo_slider = pn.widgets.IntSlider(name='Pull Request Counts', start=0, end=6000, step=50, value=0)
pulls_per_repo_slider

# Slide the pull request counter to display chart accordingly
pulls_per_repo_pipeline = (
    pulls_per_repo[pulls_per_repo.Pull_Request_Count >= pulls_per_repo_slider]
    .sort_values(by=['Pull_Request_Count'], ascending=False)
)

# Use panel to display bar chart by passing Organisation and Counts
pulls_per_repo_bar_plot = pulls_per_repo_pipeline.hvplot(kind='bar', 
                                                     x='Repository', 
                                                     y='Pull_Request_Count', 
                                                     width=600, 
                                                     height=600,
                                                     rot=90,
                                                     title='Number of Pulls per Repository')
pulls_per_repo_bar_plot
```

```python
# Number of Active vs Retired Rockets
rocket_status = (
    idf[
        (idf.Year > year_slider_min) &
        (idf.Year < year_slider_max)
    ]
    .groupby(['Rocket_Status', 'Country'])
    .size()
    .reset_index(name='Counts')
    .sort_values(by='Country')  
    .reset_index(drop=True)
)
rocket_status
```

```python
rocket_status_bar_plot = rocket_status.hvplot(kind='bar', 
                                             x='Rocket_Status',
                                             by='Country',
                                             y='Counts', 
                                             width=1000, 
                                             height=500,
                                             rot=90,
                                             title='Number of Active vs Retired Rockets')
rocket_status_bar_plot
```

```python
# Distribution of Mission Status (Success vs Failed)
mission_status = (
    idf[
        (idf.Year > year_slider_min) &
        (idf.Year < year_slider_max)
    ]
    .groupby(['Rocket_Status', 'Mission_Status', 'Country'])
    .size()
    .reset_index(name='Counts')
    .sort_values(by='Country')  
    .reset_index(drop=True)
)
mission_status_bar_plot = mission_status.hvplot(kind='bar', 
                                                     x='Mission_Status',
                                                     by='Country',
                                                     y='Counts', 
                                                     width=1000, 
                                                     height=500,
                                                     rot=90,
                                                     title='Distribution of Mission Status')
mission_status_bar_plot
```

```python
# Total Amount of Money Spent by Organisation on Space Missions
mission_price_by_country = idf.groupby(['Organisation'])['Price'].sum().to_frame().reset_index()
mission_price_by_country = mission_price_by_country[mission_price_by_country['Price'] != 0.00]
mission_price_by_country.sort_values(by='Price').reset_index(drop=True)
mission_price_by_country_table = mission_price_by_country.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width') 
mission_price_by_country_table
```

```python
launch_per_org_price = idf[['Organisation','Price']].groupby('Organisation').agg(
        LaunchCount=('Price', 'count'), 
        OrganisationperLaunch=('Price', 'mean')
    ).reset_index()
launch_per_org_price = launch_per_org_price[launch_per_org_price['OrganisationperLaunch'] != 0.00]
launch_per_org_price.sort_values(by='OrganisationperLaunch').reset_index(drop=True)
launch_per_org_price_table = launch_per_org_price.pipe(pn.widgets.Tabulator, pagination='remote', page_size = 10, sizing_mode='stretch_width')
launch_per_org_price_table
```

```python
# Chart the Number of Launches per Year

launch_per_year = (
    idf[
        (idf.Year > year_slider_min) &
        (idf.Year < year_slider_max)
    ]
    .groupby(['Year'])
    .size()
    .reset_index(name='LaunchesPerYear')
)
launch_per_year.hvplot(kind="line",x='Year', y='LaunchesPerYear')
```

```python
# Launches per Month: Which months are most popular and least popular for launches
# Some months have better weather than others. Which time of year seems to be best for space missions?
launch_per_month = (
    idf[
        (idf.Year > year_slider_min) &
        (idf.Year < year_slider_max)
    ]
    .groupby(['Month'])
    .size()
    .reset_index(name='HighestLeastLaunchPerMonth')
)
# launch_per_month.hvplot(kind="line",x='Month', y='HighestLeastLaunchPerMonth')
month_chart = launch_per_month.hvplot(kind="line",x='Month', y='HighestLeastLaunchPerMonth')
```

```python
template = pn.template.FastListTemplate(theme="dark",
    title='Space Mission Dashboard', 
    sidebar=[pn.pane.Markdown("## Settings"),   
             year_slider_min, year_slider_max],
    main=[pn.Row(pn.Column(month_chart.panel(), margin=(0,25)),
          pn.Column(launch_per_org_price_table.panel(), margin=(0,25))),
          pn.Row(pn.Column(rocket_status_bar_plot.panel(), margin=(0,25))),
          pn.Row(pn.Column(mission_status_bar_plot.panel(), margin=(0,25))),
          
         ],
    accent_base_color="#88d8b0",
    header_background="#88d8b0",
)
# template.show()
template.show();
```
